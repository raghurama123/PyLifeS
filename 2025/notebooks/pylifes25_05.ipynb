```python
import matplotlib.pyplot as plt

# Data for bar plot
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 20]

# Create bar plot
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.title('Bar Plot Example', fontsize=16)
plt.tight_layout()
plt.savefig('bar_plot_example.png')
plt.show()
```

```python
import numpy as np

# Data for histogram
data = np.random.randn(1000)

# Create histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Histogram Example', fontsize=16)
plt.tight_layout()
plt.savefig('histogram_example.png')
plt.show()
```

```python
from scipy.interpolate import make_interp_spline

# Data for spline
x = np.linspace(0, 10, 10)
y = np.sin(x)

# Smooth data
x_smooth = np.linspace(x.min(), x.max(), 3000)
spl = make_interp_spline(x, y, k=3)  # BSpline object
y_smooth = spl(x_smooth)

# Create smooth plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'o', label='Data Points')
plt.plot(x_smooth, y_smooth, label='Spline Smoothened Curve')
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.title('Spline Smoothening Example', fontsize=16)
plt.legend(fontsize=12)
plt.tight_layout()
plt.savefig('spline_smoothening_example.png')
plt.show()
```

```python
# Data for square plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create square plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sine Wave')
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.title('Square Plot Example', fontsize=16)
plt.axis('square')
plt.legend(fontsize=12)
plt.tight_layout()
plt.savefig('square_plot_example.png')
plt.show()
```

```python
import numpy as np

# Parameters for circle curve
r = 1
t = np.linspace(0, 2 * np.pi, 1000)

# circle 
x = r * np.sin(t)
y = r * np.cos(t)

# Create plot
plt.figure(figsize=(6, 6))
plt.plot(x, y, label='Circle')
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.title('Circle, parameteric plot', fontsize=16)
plt.legend(fontsize=12)
plt.tight_layout()
plt.xlim([-6,6])
plt.ylim([-6,6])
plt.savefig('cicle.png')
plt.show()
```

```python
import numpy as np

# Parameters for Lissajous curve
A = 5
B = 4
a = 10
b = 2
delta = np.pi / 2
t = np.linspace(0, 2 * np.pi, 1000)

# Lissajous curve
x = A * np.sin(a * t + delta)
y = B * np.sin(b * t)

# Create plot
plt.figure(figsize=(6, 6))
plt.plot(x, y, label='Lissajous Curve')
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.title('Lissajous Diagram Example', fontsize=16)
plt.legend(fontsize=12)
plt.tight_layout()
plt.xlim([-6,6])
plt.ylim([-6,6])
plt.savefig('lissajous_diagram_example.png')
plt.show()
```

```python
import numpy as np

# Data for contour plots
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(x, y)

# 2D Parabola
Z1 = X**2 + Y**2

# Saddle point
Z2 = X**2 - Y**2

# Create contour plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 2D Parabola contour plot
contour1 = ax1.contour(X, Y, Z1, 20, cmap='viridis')
ax1.set_title('2D Parabola', fontsize=16)
ax1.set_xlabel('X-axis', fontsize=14)
ax1.set_ylabel('Y-axis', fontsize=14)
fig.colorbar(contour1, ax=ax1)

# Saddle point contour plot
contour2 = ax2.contour(X, Y, Z2, 20, cmap='plasma')
ax2.set_title('Saddle Point', fontsize=16)
ax2.set_xlabel('X-axis', fontsize=14)
ax2.set_ylabel('Y-axis', fontsize=14)
fig.colorbar(contour2, ax=ax2)

plt.tight_layout()
plt.savefig('contour_plots_example.png')
plt.show()
```
