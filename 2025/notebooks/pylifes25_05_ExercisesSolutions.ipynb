### **Exercise 10:**  
Write a function `compute_rmsd(atom_list1, atom_list2)` that:
1. Takes two lists of atomic coordinates.
2. Computes and returns the Root Mean Square Deviation (RMSD) between them.

```python
import numpy as np

def compute_rmsd(atom_list1, atom_list2):
    diff = np.array(atom_list1) - np.array(atom_list2)
    return np.sqrt(np.mean(np.sum(diff**2, axis=1)))
```

## **Advanced Challenges**

### **Challenge 1:**  
Write a function `generate_pdb(num_residues)` that:
1. Creates a random PDB-like file with `num_residues` amino acids.
2. Saves it as `random_structure.pdb`.

```python
import random

def generate_pdb(num_residues):
    residues = ['ALA', 'GLY', 'LEU', 'LYS', 'VAL']
    with open('random_structure.pdb', 'w') as file:
        for i in range(1, num_residues + 1):
            res = random.choice(residues)
            file.write(f"ATOM      {i}  CA  {res} A   {i}     {random.uniform(0, 10):.3f} {random.uniform(0, 10):.3f} {random.uniform(0, 10):.3f}  1.00 20.00           C\n")
    print("PDB file 'random_structure.pdb' created.")
```
