```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Step 1: Simulate mRNA synthesis (sequence generation)
def generate_mrna_sequence(length=30):
    bases = ['A', 'U', 'C', 'G']
    return ''.join(random.choice(bases) for _ in range(length))

# Step 2: Apply nucleoside modifications (replace 'U' with pseudouridine )
def modify_mrna_sequence(sequence):
    return sequence.replace('U', '')

# Step 3: Simulate immune response (activation scores)
def simulate_immune_response(sequence, modified=False):
    # Higher immune response for unmodified sequences due to stronger recognition
    response_factor = 0.5 if modified else 1.0
    return np.random.normal(loc=50 * response_factor, scale=10, size=len(sequence))

# Generate original and modified mRNA sequences
original_mrna = generate_mrna_sequence()
modified_mrna = modify_mrna_sequence(original_mrna)

# Simulate immune responses
original_response = simulate_immune_response(original_mrna)
modified_response = simulate_immune_response(modified_mrna, modified=True)

# Plot results
plt.figure(figsize=(12, 6))
plt.hist(original_response, bins=15, alpha=0.7, label='Unmodified mRNA', color='red')
plt.hist(modified_response, bins=15, alpha=0.7, label='Modified mRNA ()', color='blue')
plt.axvline(np.mean(original_response), color='red', linestyle='--', label='Mean Response (Unmodified)')
plt.axvline(np.mean(modified_response), color='blue', linestyle='--', label='Mean Response (Modified)')

plt.title('Simulated Immune Response to mRNA Sequences')
plt.xlabel('Immune Activation Score')
plt.ylabel('Frequency')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Print sequences
print("Original mRNA:", original_mrna)
print("Modified mRNA:", modified_mrna)
```

Output

    Histogram: The immune response for unmodified mRNA will typically be higher (indicating stronger activation) than that for modified mRNA, illustrating the reduced immune response due to pseudouridine incorporation.
    Sequences: The original and modified mRNA sequences will be printed for comparison.

This simulation showcases the impact of nucleoside modifications on reducing immune activation, a crucial step in enabling the success of mRNA vaccines.

```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Simulate mRNA encoding a spike protein
def generate_mrna(length=30):
    codons = ['AUG', 'UUU', 'GGG', 'CCC', 'AAA', 'UGA']  # Example codons
    return [np.random.choice(codons) for _ in range(length)]

# Step 2: Simulate translation of mRNA to spike protein
def translate_mrna_to_protein(mrna_sequence):
    protein = [codon.replace('AUG', 'Methionine')
                     .replace('UUU', 'Phenylalanine')
                     .replace('GGG', 'Glycine')
                     .replace('CCC', 'Proline')
                     .replace('AAA', 'Lysine')
                     .replace('UGA', 'Stop') for codon in mrna_sequence]
    return protein

# Step 3: Simulate immune system response
def immune_response(protein_sequence):
    antibody_production = [np.random.randint(10, 100) for _ in protein_sequence if 'Stop' not in _]
    return sum(antibody_production), antibody_production

# Generate an mRNA sequence
mrna_sequence = generate_mrna(20)

# Translate the mRNA into a spike protein
protein_sequence = translate_mrna_to_protein(mrna_sequence)

# Simulate the immune response
total_antibodies, antibody_distribution = immune_response(protein_sequence)

# Plot the antibody response
plt.figure(figsize=(10, 5))
plt.bar(range(len(antibody_distribution)), antibody_distribution, color='blue', alpha=0.7)
plt.axhline(total_antibodies / len(antibody_distribution), color='red', linestyle='--', label='Average Response')
plt.title('Immune Response to mRNA Vaccine')
plt.xlabel('Protein Segment')
plt.ylabel('Antibody Production Level')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Output results
print("mRNA Sequence:", ' '.join(mrna_sequence))
print("Translated Protein Sequence:", ' -> '.join(protein_sequence))
print(f"Total Antibodies Produced: {total_antibodies}")
```

How it Works:

    mRNA Generation: The generate_mrna() function creates a sequence of codons representing genetic instructions.
    Translation: The translate_mrna_to_protein() function mimics ribosomes translating mRNA into spike proteins.
    Immune Response: The immune_response() function simulates antibody production levels in response to each protein segment.

Output:

    Bar Chart: Shows antibody production levels for each segment of the protein, highlighting immune system activation.
    Textual Output: Displays the generated mRNA sequence, translated protein sequence, and total antibody production.

This code visualizes how mRNA vaccines teach the immune system to recognize and respond to a virus, aligning with the key discoveries of Karik and Weissman.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Parameters
num_samples = 100  # Number of immune response measurements

# Simulate immune response data
np.random.seed(42)  # For reproducibility
unmodified_response = np.random.normal(loc=80, scale=10, size=num_samples)  # High activation
modified_response = np.random.normal(loc=60, scale=8, size=num_samples)  # Lower activation due to pseudouridine

# Perform statistical analysis (t-test)
t_stat, p_value = stats.ttest_ind(unmodified_response, modified_response)

# Print statistical results
print("Immune Response Statistics:")
print(f"Mean (Unmodified mRNA): {np.mean(unmodified_response):.2f}")
print(f"Mean (Modified mRNA): {np.mean(modified_response):.2f}")
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Visualization
plt.figure(figsize=(14, 6))

# Boxplots for immune response comparison
plt.subplot(1, 2, 1)
plt.boxplot([unmodified_response, modified_response], labels=["Unmodified mRNA", "Modified mRNA"], patch_artist=True,
            boxprops=dict(facecolor="lightblue", color="blue"),
            medianprops=dict(color="red"))
plt.title("Boxplot of Immune Response")
plt.ylabel("Immune Activation Score")

# Histograms for immune response distribution
plt.subplot(1, 2, 2)
plt.hist(unmodified_response, bins=15, alpha=0.7, label="Unmodified mRNA", color="red")
plt.hist(modified_response, bins=15, alpha=0.7, label="Modified mRNA", color="blue")
plt.axvline(np.mean(unmodified_response), color='red', linestyle='--', label='Mean (Unmodified)')
plt.axvline(np.mean(modified_response), color='blue', linestyle='--', label='Mean (Modified)')
plt.title("Histogram of Immune Response")
plt.xlabel("Immune Activation Score")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()
```
